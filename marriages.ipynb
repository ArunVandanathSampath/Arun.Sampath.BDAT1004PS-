Create a line plot showing the number of marriages and divorces per capita in the U.S. between 1867 and 2014. Label both lines and show the legend.
Don't forget to label your axes!

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file into a DataFrame
data = pd.read_csv("us-marriages-divorces-1867-2014.csv")

# Create a new column for marriages per capita
data["Marriages_per_capita"] = data["Marriages_per_1000"] / 10

# Create a new column for divorces per capita
data["Divorces_per_capita"] = data["Divorces_per_1000"] / 10

# Create a line plot of marriages and divorces per capita
plt.plot(data["Year"], data["Marriages_per_capita"], label="Marriages per capita")
plt.plot(data["Year"], data["Divorces_per_capita"], label="Divorces per capita")

# Set the title and labels for the plot
plt.title("Marriages and Divorces per Capita in the U.S.")
plt.xlabel("Year")
plt.ylabel("Number per 1000 people")

# Show the legend
plt.legend()

# Show the plot
plt.show()
```
